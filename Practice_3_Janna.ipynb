```python
#1-тапсырма
import pandas as pd

df = pd.read_csv('Salaries.csv', nrows=10)

data_rows = df.values.tolist()

for row in data_rows:
    print('. '.join(map(str, row)))
print("---")
print(f'Тізімнің ұзындығы: {len(data_rows)}')
print("---")
print(f'Тізімнің алғашқы элементі:\n{data_rows[0]}')
```

```python
#2-тапсырма

first_three_names_list = [inner_list[1] for inner_list in data_rows[:3]]
first_three_jobtitle_list = [inner_list[2] for inner_list in data_rows[:3]]

first_three_names = tuple(first_three_names_list)
first_three_jobtitle = tuple(first_three_jobtitle_list)

print(f"Аты-жөні(кортеж):\n{first_three_names}\n\nЖұмысы атағы(кортеж):\n{first_three_jobtitle}")

keys_tuple = (first_three_names)
zipped_pairs = zip(keys_tuple, first_three_jobtitle)
my_dictionary = dict(zipped_pairs)

print(f'\nСөздік:\n{my_dictionary}')
```

```python
#3-тапсырма
base_pay = [str(inner_list[3]) for inner_list in data_rows]

print(f'Қызметкерлердің базалық жалақысы:\n{base_pay}')
```

```python
#4-тапрсырма
unique_titles = set(inner_list[2] for inner_list in data_rows)
print('Қызметкерлердің жұмыс атағы:')
for element in unique_titles:
    print(element)
```

Қызметкерлердің жұмыс атағын жиын арқылы алған тиімдірек. Себебі, қайталанатын атаулар кездеспейді.
