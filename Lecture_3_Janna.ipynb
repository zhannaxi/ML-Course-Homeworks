```python
import json

data = {
    'students': [
        {'name': 'Nazym', 'grade': 85},
        {'name': 'Arman', 'grade': 92},
        {'name': 'Adiya', 'grade': 96}
    ],
    'course': 'Data Mining',
}

with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

with open('data.json', 'r', encoding='utf-8') as f:
    loaded = json.load(f)
print('Loaded:', loaded)

average_grade = sum(student['grade'] for student in data['students']) / len(data['students'])
print('Average Grade:', average_grade)
```
